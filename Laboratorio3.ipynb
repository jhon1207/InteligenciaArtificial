```python
# Zu√±iga Mendez Jhonny Agustin
# datos = estado
class Nodo:
    def __init__(self, datos, hijo=None):
        self.datos = datos
        self.hijos = []
        self.padre = None
        self.costo = None
        self.set_hijo(hijo)
        
    def set_hijo(self, hijo):
        if (hijo is not None):
            self.hijos.append(hijo)
            if self.hijos is not None:
                for h in self.hijos:
                    h.padre = self
                
    def get_hijos(self):
        return self.hijos
    
    def set_padre(self, padre):
        self.padre = padre
        
    def get_padre(self):
        return self.padre

    def set_datos(self, datos):
        self.datos = datos
    
    def get_datos(self):
        return self.datos
    
    def set_costo(self, costo):
        self.costo = costo
        
    def get_costo(self):
        return self.costo
    
    def equal(self, nodo):
        if self.get_datos() == nodo.get_datos():
            return True
        else:
            return False
    
    def en_lista(self, lista_nodos):
        enlistado = False
        for n in lista_nodos:
            if self.equal(n):
                enlistado = True
        return enlistado
    
    def __str__(self):
        return str(self.get_datos())
[4]
def bpa(estado_inicio, estado_solucion):
    resuelto = False
    nodos_visitados = []
    nodos_frontera = []
    nodo_inicio = Nodo(estado_inicio)
    nodos_frontera.append(nodo_inicio)
    
    while resuelto == False and len(nodos_frontera) != 0:
        #Al iniciar el pop en 0 se convierte en una lista FIFO ya que con esto se estaria convirtiendo
        #en una cola por que con esto estaria sacando el primer nodo que entro
        #nodo_actual = nodos_frontera.pop(0) 
        #Para que se convierta en una lista LIFO se tiene que iniciar el pop vacio y con esto estaria 
        #sacando el ultimo nodo y se convierte en una pila y por lo tanto seria una busqueda primero 
        #en profundidad
        nodo_actual = nodos_frontera.pop() 
        nodos_visitados.append(nodo_actual)
        if nodo_actual.get_datos() == estado_solucion:
            resuelto = True
            return nodo_actual
        else:
            #Para generar nodos hijos para n caracteres se utiliza este primer for donde hace que i va ir
            #creciendo dependiendo la longitud del estado inicial menos 1 y con eso se genera los nodos de capa en capa
            for i in range(len(estado_inicio)-1): 
                #El segundo for tambien depende de la longitud del estado inicial menos 1 para que esto nos genere en cada
                #movimiento vaya de uno en uno dependiendo de la longitud
                for j in range(len(estado_inicio)-1): 
                    hijo_datos = nodo_actual.get_datos().copy()    
                    temp = hijo_datos[j]
                    hijo_datos[j] = hijo_datos[j+1]
                    hijo_datos[j+1] = temp
                    hijo = Nodo(hijo_datos)
                    #print(f'hijo {j+1}',hijo_datos)
                    if not hijo.en_lista(nodos_visitados) and not hijo.en_lista(nodos_frontera):
                        nodo_actual.set_hijo(hijo)
                        nodos_frontera.append(hijo)
```
